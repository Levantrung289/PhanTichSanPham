```python
import pandas as pd
```

```python
#read data
df = pd.read_csv('df_product.csv')
df.head()
```

```python
# bỏ 2 cột dư
df.drop(['Unnamed: 0.1','Unnamed: 0'], axis = 1,inplace = True)
```

```python
df.head()
```

```python
# bỏ các kí tự thừa trong giá tiền
l_price = df["Price"].apply(lambda x: x.replace('₫','',2).replace('.','',5))
l_price = l_price.apply(lambda x: x.split('\n'))
#l_price = l_price.apply(lambda x: )
l_price
```

```python
# tách giá trị của cột price thành 2 phần
for element in l_price.values:
  if len(element) == 2:
    pass
  else:
    element.append(element[0])
l_price
```

```python
#lấy sold price và init price
sold_price = []
init_price = []
for index in range(len(l_price)):
  if l_price[index][0].find(' - ') != -1:
    sold_price.append(l_price[index][1].split(' - ')[1])
  else:
    sold_price.append(l_price[index][1])

  if l_price[index][1].find(' - ') != -1:
    init_price.append(l_price[index][0].split(' - ')[1])
  else:
    init_price.append(l_price[index][0])
```

```python
# thêm 2 cột vào data frame
df.insert(3,column = 'Sold_price', value = sold_price)
df.insert(4,column = 'Init_price', value = init_price)
df.head()
```

```python
#bỏ cột price
df.drop('Price',axis = 1, inplace = True)
df.head()
```

```python
# bỏ các kí tự rác trong cột sold_num
Sold_num = df['Sold_num'].apply(lambda x: str(x).replace(',','').replace('k','').replace('Đã bán ','').replace('nan',''))
df.drop('Sold_num',axis = 1, inplace = True)
df.insert(5,column = 'Sold_num', value = Sold_num)
df.head()
```

```python
df.info()
```

```python
#chuyển các cột về kiểu dữ liệu float
df['Sold_price'] = pd.to_numeric(df['Sold_price'])
df['Init_price'] = pd.to_numeric(df['Init_price'])
df['Sold_num'] = pd.to_numeric(df['Sold_num'])
```

```python
# tính discount của sản phẩm
discount = 1.0 - (df['Sold_price'] / df['Init_price'])
```

```python
#thêm cột discount vào data frame
df.insert(6,column = 'Discount', value = discount)
df.head()
```
